3. Les données sont-elles stationnaires ?

Pour vérifier la stationnarité d'une série temporelle, on peut utiliser le test de Dickey-Fuller augmenté (ADF Test). Ce test a pour hypothèse nulle que la série temporelle peut être représentée par une racine unitaire, c'est-à-dire qu'elle n'est pas stationnaire. Si le résultat du test est inférieur au seuil de signification (par exemple, 0.05), on rejette l'hypothèse nulle et on peut conclure que la série est stationnaire.

```python
from statsmodels.tsa.stattools import adfuller

def test_stationnarite(df):
    results = {}
    for column in df.columns:
        test_result = adfuller(df[column])
        results[column] = {'Test Statistic': test_result[0], 'p-value': test_result[1]}
        print(f"{column}: Statistique de test = {test_result[0]:.4f}, p-value = {test_result[1]:.4f}")
        if test_result[1] < 0.05:
            print(f"--> La série {column} est stationnaire.\n")
        else:
            print(f"--> La série {column} n'est pas stationnaire.\n")
    return results

stationnarite_results = test_stationnarite(features_df)
print(stationnarite_results['t2m'])
```

13. Entraînez les modèles sélectionnés sur le dataset d’entraînement préparé et prétraité.
L’utilisation d’un dictionnaire et d’une boucle sur les clés est conseillé.

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix, ConfusionMatrixDisplay
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
import matplotlib.pyplot as plt

models = {
    'Logistic Regression': LogisticRegression(max_iter=1000),
    'Random Forest': RandomForestClassifier(n_estimators=100),
    'XGBoost': XGBClassifier(use_label_encoder=False, eval_metric='logloss')
}

for name, model in models.items():
    print(f'--- {name} ---')

    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)

    print(f'Accuracy: {accuracy:.4f}')
    print(f'Precision: {precision:.4f}')
    print(f'Recall: {recall:.4f}')

    cm = confusion_matrix(y_test, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm)
    disp.plot()
    plt.title(f'Matrice de confusion pour {name}')
    plt.show()
```

```python
#for XGBOOST
from xgboost import XGBClassifier
param_grid_xgb = {
    'n_estimators': [100, 200, 300],
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 4, 5],
    'subsample': [0.8, 0.9, 1.0],
    'colsample_bytree': [0.8, 0.9, 1.0]
}

grid_search_xgb = GridSearchCV(XGBClassifier(use_label_encoder=False, eval_metric='logloss'), param_grid_xgb, cv=5, scoring='recall', n_jobs=-1)
grid_search_xgb.fit(X_train, y_train)

print("Meilleurs paramètres (XGBoost):", grid_search_xgb.best_params_)
print("Meilleure recall (XGBoost):", grid_search_xgb.best_score_)
```

Dans scikit-learn GridSearch permet de tester systématiquement plusieurs combinaisons d'hyperparamètres en spécifiant une "grille" de valeurs possibles pour chaque hyperparamètre. La meilleure combinaison est celle qui donne la meilleure performance sur les données de validation croisée. Ici nous voulons la combinaison qui maximise notre Recall.

15. Présentez les résultats de ces métriques en introduisant une quantification de l’incertitude
associée (e.g. K-fold).


```python
from sklearn.model_selection import cross_val_score, StratifiedKFold
from sklearn.metrics import make_scorer, accuracy_score, recall_score
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier

models = {
    'Logistic Regression': LogisticRegression(max_iter=1000),
    'Random Forest': RandomForestClassifier(n_estimators=100, max_depth=10, min_samples_leaf=1, min_samples_split=10),
    'XGBoost': XGBClassifier(use_label_encoder=False, eval_metric='logloss', colsample_bytree=0.8, learning_rate=0.2, max_depth=3, n_estimators=100, subsample=1.0)
}

cv = StratifiedKFold(n_splits=5)

results = {}

for name, model in models.items():
    cv_accuracy = cross_val_score(model, X_train, y_train, cv=cv, scoring='accuracy')
    cv_recall = cross_val_score(model, X_train, y_train, cv=cv, scoring='recall')

    results[name] = {
        'Accuracy': {'Mean': cv_accuracy.mean(), 'Std': cv_accuracy.std()},
        'Recall': {'Mean': cv_recall.mean(), 'Std': cv_recall.std()}
    }

    print(f"{name} - Accuracy: {cv_accuracy.mean():.4f} (±{cv_accuracy.std():.4f}), Recall: {cv_recall.mean():.4f} (±{cv_recall.std():.4f})")
```

Le meilleur Modèle est donc le XGBOOST car il est celui nous donnant le meilleur recall et la meilleur accuracy
